```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\ACER\Downloads\cars-1.csv")
```

```python
import matplotlib.pyplot as plt
# Plotting
plt.figure(figsize=(10, 6))  # Define the figure size (width, height)
# Plotting the line plot
plt.plot(df['model'], df['hp'], marker='o', linestyle='-', color='b', label='Horsepower')
# Adding labels and title
plt.xlabel('Models of the cars')  # X-axis label
plt.ylabel('Horse-Power of Cars')  # Y-axis label
plt.title('Model Names vs HorsePower')  # Plot title
# Rotating x-axis labels for better readability (if needed)
plt.xticks(rotation=45)
# Adding grid lines
plt.grid(True)
# Adding legend
plt.legend()
# Display the plot
plt.tight_layout()  # Ensures all labels and ticks are visible properly
plt.show()
```
