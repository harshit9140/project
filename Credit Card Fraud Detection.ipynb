```python
#6 Split the dataset in train and test set and have a 70:30 split ratio for the model.
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv(r"C:\Users\ACER\Downloads\creditcard.csv")
```

```python
data
```

```python
X = data.drop(columns=['Class'])
y = data['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
train_set = pd.concat([X_train, y_train], axis=1)
test_set = pd.concat([X_test, y_test], axis=1)
```

```python
train_set.to_csv('train_data.csv', index=False)
test_set.to_csv('test_data.csv', index=False)
```

```python
print(f"Train set size: {train_set.shape}")
print(f"Test set size: {test_set.shape}")
```

```python
#7. Now use a decision tree and random forest model for training on top of the train set.
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
dt_model = DecisionTreeClassifier(random_state=42)
```

```python
dt_model.fit(X_train, y_train)
```

```python
rf_model = RandomForestClassifier(random_state=42)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred_dt = dt_model.predict(X_test)
y_pred_rf = rf_model.predict(X_test)
```

```python
# Evaluating the Decision Tree model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Decision Tree Model")
print("Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_dt))
print("Classification Report:\n", classification_report(y_test, y_pred_dt))
```

```python
# Evaluating the Random Forest model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Random Forest Model")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
```

```python
#8 compare the predictions of both models using predict().
comparison = pd.DataFrame({
    'Actual': y_test,
    'Decision Tree': y_pred_dt,
    'Random Forest': y_pred_rf
})

print("\nComparison of predictions:\n")
print(comparison.head())
```

```python
# Checking how often the predictions agree
agreement = (y_pred_dt == y_pred_rf)
print("\nNumber of agreeing predictions:", np.sum(agreement))
print("Number of disagreeing predictions:", np.sum(~agreement))
```

```python
#9. Compare the accuracy of both models using score().
accuracy_dt = dt_model.score(X_test, y_test)
accuracy_rf = rf_model.score(X_test, y_test)

print("\nAccuracy Comparison using score() method:")
print(f"Decision Tree Accuracy: {accuracy_dt}")
print(f"Random Forest Accuracy: {accuracy_rf}")
```
